```python
#Importing libraries
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('firms.csv',skipinitialspace=True, encoding= 'unicode_escape')
```

### Data prerocessing

```python
#dropping REGION, COUNTRY and CITY as it doesn't give particular information
df = data.drop(columns = ['HQ REGION','HQ COUNTRY', 'HQ CITY','TYPE','LINKEDIN'])
```

```python
N = len(df)
df
```

```python
#Making the feature LAUNCH DATE binary (either before 1990 or after 1990)

date = [int(re.search('(\d+)', df['LAUNCH DATE'][i]).group()) for i in range(N)]
df['LAUNCH DATE'] = [1 if date[i] > 1989 else 0 for i in range(N)]
```

```python
#Creating lists to store the labelled samples

org_gov = []
mature_comp = []
startup = []
unclassified= []
univ = []
```

```python
#Lowercasing the strings and replacing Not A Number (NaN) values by empty strings.

df['TAGS'].fillna('',inplace = True) 
df['TAGLINE'].fillna('',inplace = True) 
df['GROWTH STAGE'].fillna('',inplace = True) 
df['TAGS'] = [df['TAGS'][i].lower().replace('-',' ') for i in df.index.values]
df['TAGLINE'] = [df['TAGLINE'][i].lower().replace('-',' ') for i in df.index.values]
```

```python
#First Node of the tree

for i in df.index.values : 
    
    if df['WEBSITE'][i].find('.org') >= 0: 
        org_gov.append(i)
        continue
        
    elif df['WEBSITE'][i].find('.gov') >= 0 :
        org_gov.append(i)
        continue
        
    elif (df['TAGLINE'][i].find('profit') >= 0 and df['TAGLINE'][i].find('profitab') == -1 ):
        org_gov.append(i)
        continue

#Visualizing the first extracted class "Government/Non-Profit"
df.iloc[org_gov,:]
```

```python
df.drop(org_gov, 0, inplace = True)
```

```python
#Second Node of the Tree

df['NAME'] = [df['NAME'][i].lower() for i in df.index.values]

for i in df.index.values : 
    
    if df['NAME'][i].find('school') >= 0: 
        univ.append(i)
        continue
        
    elif df['NAME'][i].find('universit') >= 0 :
        univ.append(i)
        continue
        
#Visualizing the second extracted class "Government/Non-Profit"
df.loc[univ,:]
```

```python
df.drop(univ, 0, inplace = True)
```

```python
#Splitting on the attribute LAUNCH DATE 

new = df[df['LAUNCH DATE'] == 1]   #created after 1990
old = df[df['LAUNCH DATE'] == 0]   #created before 1990
```

```python
for i in old.index.values : 
    
    if old['TAGLINE'][i].find('lead') >= 0 or old['TAGLINE'][i].find('international') >= 0 or old['TAGLINE'][i].find('global') >= 0 or old['TAGLINE'][i].find('world') >= 0: 
        mature_comp.append(i)
        continue
        
    elif old['GROWTH STAGE'][i].find('late growth stage') >= 0 :
        mature_comp.append(i)
        continue
    else : 
        unclassified.append(i)
```

```python
df.drop(mature_comp + unclassified, 0, inplace = True)
```

```python
#creating the feature TECH (1 if it is a Tech company, 0 if not)

#Vocabulary associated to Tech Companies
keywords = ['tech', ' it ', 'digital', 'platform', 'app', 'intelligence', 'software', 'data', 'web','mobile', ' e ',' ai ',  'data']
tech = []

for i in df.index.values :
    
    found = False
    for word in keywords:
        if df['TAGLINE'][i].find(word) >= 0 or df['TAGS'][i].find(word) >= 0 or df['NAME'][i].find(word) >= 0  :    
            found = True      
    if found :
        tech.append(1)
    else :
        tech.append(0)
```

```python
df['tech'] = tech 

#Splitting on the attribute TECH 

tech_comp = df[df['tech'] == 1]   #tech company
non_tech = df[df['tech'] == 0]   #else
```

```python
#Vocabulary associated to Mature Companies
keys = ['lead','international', 'global', 'world']

for i in non_tech.index.values : 
    
    if non_tech['GROWTH STAGE'][i] == 'late growth stage' :
        mature_comp.append(i)
        continue
    
    found = False
    
    for word in keys:
        if non_tech['TAGLINE'][i].find(word) >= 0 or non_tech['TAGS'][i].find(word) >= 0 :    
            found = True     
            
    if found :
        mature_comp.append(i)
        
    else : 
        unclassified.append(i)
```

```python
for i in tech_comp.index.values : 
    
    if tech_comp['GROWTH STAGE'][i] == 'seed stage' or tech_comp['GROWTH STAGE'][i] == 'early stage' : 
        startup.append(i)
        continue
        
    found = False
    
    for word in keys:
        if tech_comp['TAGLINE'][i].find(word) >= 0 or tech_comp['TAGS'][i].find(word) >= 0 :    
            found = True     
            
    if found :
        mature_comp.append(i)
        
    else : 
        startup.append(i)
```

```python
#verifying that all instances have been classified
print(len(unclassified + mature_comp + univ + org_gov + startup) == N)

#counting the number of elements from each class
print(len(unclassified), len(mature_comp), len(univ), len(org_gov), len (startup))
```

```python
TYPE = [0]*N
for i in unclassified:
    TYPE[i] = 'Unclassified'
for i in mature_comp:
    TYPE[i] = 'Mature Company'
for i in univ:
    TYPE[i] = 'Universities/Schools'
for i in org_gov:
    TYPE[i] = 'Government/Non-profit'
for i in startup:
    TYPE[i] = 'Startup'
```

```python
data['TYPE'] = TYPE
```
